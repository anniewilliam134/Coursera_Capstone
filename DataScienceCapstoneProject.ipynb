## Segmenting and Clustering Neighborhoods in Toronto

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
#Used https://www.youtube.com/watch?v=OXA_ZD1gR6A as reference to learn to extract table from website
wikitables = read_html(url,attrs={"class":"wikitable"})
```

```python
#Verify table was extracted
print("Extracted {num} wikitables".format(num=len(wikitables)))
```

```python
#Create df for this table
Toronto=wikitables[0]
```

```python
#Drop any values that have "Not assigned" as a Neighbourhood
TorontoTable = Toronto[~Toronto.Borough.str.contains("Not assigned")]
```

```python
#Set Neighbourhood name to Bourough Name if Neighbourhood name is "Not assigned"
TorontoTable.loc[TorontoTable['Neighbourhood'] == 'Not assigned', 'Neighbourhood'] = TorontoTable['Borough']
```

```python
TorontoTable.head(20)
```

```python
#Format so if Postcode and Bourough have more than one neighbourhood, append and join with a comma
#used example from https://stackoverflow.com/questions/36271413/pandas-merge-nearly-duplicate-rows-based-on-column-value
CleanedTT = TorontoTable.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
```
